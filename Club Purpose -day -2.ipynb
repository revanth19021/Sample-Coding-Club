```python
# elif ladder
age=int(input("enter age: "))
if(age==18):
    print("Minor")
elif(age>18 and age<=25):
    print("Youth")
elif(age>25 and age<=35):
    print("Next generation")
else:
    print("he is above 35")
```

```python
age=int(input("Enter age: "))
if(age>18):
    if(age>=18 and age<=25):
        print("Youth")
    elif(age>25 and age<=35):
        print("Next Generation")
    else:
        print("Above 35")
else:
    print("Enter valid input")
```

```python
# Nested if
if(condition 1):
    if(condition 2):
        print("Statement1")
    elif(condition 3):
        print("Elif block")
    else:
        print("Statement2")
else:
    print("Else block")
```

```python
# monday - saturday
# After 3.10 version this is included

value=input("Enter the number: ")
match value:
    case '1':print("Monday")
    case '2':print("Tuesday")
    case '3':print("Wednesday")
    case '4':print("Thursday")
    case '5':print("Friday")
    case '6':print("Saturday")
    case '-':print("Enter valid input:")
```
