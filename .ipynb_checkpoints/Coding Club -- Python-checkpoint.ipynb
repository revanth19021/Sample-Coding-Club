```python
# Palindrome reverse of a number 
# 121 = 121
# 323=323
# 141 =141

a=int(input("Enter a number: "))
temp=a
sum1=0
while temp>0:
    rem=temp%10
    sum1=sum1*10+rem
    temp=temp//10
return sum1==a
# 4th iteration
# sum1=121 temp=0
# while temp>0 > false

```

```python
# Calculate factorial of a given number
# take one input as a variable
n=int(input())
fact=1
for i in range(1,n+1):
    fact=fact*i
print(fact)
```
